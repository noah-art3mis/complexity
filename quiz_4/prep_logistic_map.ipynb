```python
import matplotlib.pyplot as plt

def logistic_map(_lambda, x):
    return _lambda * x * (1 - x)
```

```python
def make_map(_lambda, _range, init=0.5):
    xn = []
    xn.append(logistic_map(_lambda, init))

    for n in range(1, _range):
        xn.append(logistic_map(_lambda, xn[n-1]))
    
    return xn
```

```python
RANGE = 50

x = list(range(RANGE))

vars = [2, 3]
for i in vars:
    plt.plot(x,make_map(i, RANGE, init=0.1), label=str(i))

plt.legend()
plt.title('different rates');
```

```python
RANGE = 20

x = list(range(RANGE))

vars = [0.5, 0.9, 0.1]
for i in vars:
    plt.plot(x,make_map(1, RANGE, init=i), label=str(i))
    plt.plot(x,make_map(3.5, RANGE, init=i), label=str(i))
    

plt.legend()
plt.title('different inits');
```

```python
import matplotlib.pyplot as plt
import numpy as np

m = 1000  # Carrying capacity
k = 1     # Growth rate
_range = 10000  # Number of iterations

def logistic_equation(k, p, m):
    return k * p * (1 - (p/m))

init = 3  # Initial population
map = [init]  # Start with the initial value

# Iterate to generate the map
for i in range(_range):
    map.append(logistic_equation(k, map[-1], m))  # Use the last value in the list

# Plot the results
plt.plot(range(_range + 1), map)
plt.xlabel("Time Step")
plt.ylabel("Population")
plt.title("Logistic Map")
plt.show()
```
