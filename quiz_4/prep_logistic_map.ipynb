```python
def logistic_map(_lambda, x):
    return _lambda * x * (1 - x)
```

```python
def make_map(_lambda, _range, init=0.5):
    xn = []
    xn.append(logistic_map(_lambda, init))

    for n in range(1, _range):
        xn.append(logistic_map(_lambda, xn[n-1]))
    
    return xn
```

```python
from sre_constants import RANGE
import matplotlib.pyplot as plt

RANGE = 100

x = list(range(RANGE))

plt.plot(x,make_map(1, RANGE), label='m1')
plt.plot(x,make_map(0.5, RANGE), label='m2')
plt.plot(x,make_map(2, RANGE), label='m3')
plt.plot(x,make_map(3.8, RANGE), label='m4')
plt.legend();
```
